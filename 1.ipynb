```python
%%manim -ql ContinuedRoots
from manim import *

config.background_color = "#000025"

class ContinuedRoots(Scene):
    def construct(self):

        # Correct LaTeX for y = x + sqrt{x + sqrt{y + sqrt{x + ...}}}
        expr = MathTex(
            r"y = \sqrt{x + \sqrt{y + \sqrt{x + \cdots}}}"
        )

        quest = MathTex(r"Find\;",
            r"y\;=\;\frac{dy}{dx}"
        ).next_to(expr, DOWN)

        colgrp = VGroup(expr, quest)
        colgrp.set_color(YELLOW)

        self.play(Write(expr))
        self.wait(0.5)
        self.play(Write(quest))
        self.play(colgrp.animate.to_edge(UP))

        # Step 1 substitution
        st1 = MathTex(
            r"y\;=\;",r"\sqrt{x + \sqrt{y + \sqrt{x + \cdots}}}"
        )
        self.play(Write(st1))

        # Step 2 substitution
        text1 = MathTex(
            r"\text{Substituting }",
            r"y = \sqrt{x + \sqrt{y + \sqrt{x + \cdots}}}"
        )
        text1.next_to(st1,DOWN).set_color("#FF4EC4")
        self.play(Write(text1))
        rect = SurroundingRectangle(text1[1],color=GREEN)
        self.play(Create(rect))
        self.wait(2)
        rect2 = SurroundingRectangle(st1[1],color=GREEN)
        self.play(ReplacementTransform(rect,rect2))
        self.wait(1)
        st2 = MathTex(r"y =",r" \sqrt{x + \sqrt{y + y}}")
        rect3 = SurroundingRectangle(st2[1],color=GREEN)
        self.play(ReplacementTransform(st1,st2),ReplacementTransform(rect2,rect3))
        self.wait()
        self.play(FadeOut(rect3))
        st3 = MathTex(r"y = \sqrt{x + \sqrt{2y}}")
        self.play(ReplacementTransform(st2,st3),FadeOut(text1))
        self.wait(1)
        text2 = MathTex(r"\text{Now differentiating both sides w.r.t x}")
        st4 = MathTex(r"y^2 = x + \sqrt{2y}")
        self.play(ReplacementTransform(st3,st4))
        st5 = MathTex(r"y^2 - x = \sqrt{2y}")
        self.play(ReplacementTransform(st4,st5))
        st6 = MathTex(r"(y^2 - x)^2 = 2y")
        self.play(ReplacementTransform(st5,st6))
        st7 = MathTex(r"y^4 - 2xy^2 + x^2 = 2y")
        self.play(ReplacementTransform(st6, st7))
        text2.next_to(st7,DOWN).set_color("#FF4EC4")
        self.play(Write(text2))
        self.wait(0.5)
        self.play(FadeOut(text2))
        self.wait(1)
        stt = Text(
    "Now, By using Chain Rule and Quotient Rule",
    t2c={
        "Now, By using ": "#ff4ec4",
        "Chain Rule": "#4fe7ff",
        " and ": "#ff4ec4",
        "Quotient Rule": "#4fe7ff",
    },font_size =30
)
        stt.next_to(st7,DOWN)
        self.play(Write(stt))
        self.wait(0.5)
        self.play(FadeOut(stt))
        
```
